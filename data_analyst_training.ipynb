````python
# ▶️ SETUP 2/2 — the self-grading engine (run me second)
# Defines: check_answer, check_quiz, record/progress, show_hint, show_solution.
# These NEVER raise on a wrong/empty answer — they print ❌ and a hint instead,
# so the whole notebook always runs top-to-bottom cleanly.

progress = {}  # key -> {"passed": bool, "earned": int, "max": int}


def record(key, passed, points=1):
    progress[key] = {"passed": bool(passed), "earned": points if passed else 0, "max": points}


def _equal(a, b):
    """Type-aware, exception-safe equality for ints/floats/lists/dicts/np/pandas."""
    try:
        if a is None:
            return False
        if isinstance(a, (pd.DataFrame, pd.Series)) or isinstance(b, (pd.DataFrame, pd.Series)):
            try:
                return bool(a.equals(b))
            except Exception:
                return False
        if isinstance(a, np.ndarray) or isinstance(b, np.ndarray):
            try:
                return bool(np.allclose(np.asarray(a, dtype=float), np.asarray(b, dtype=float)))
            except Exception:
                try:
                    return bool(np.array_equal(np.asarray(a), np.asarray(b)))
                except Exception:
                    return False
        if isinstance(a, bool) or isinstance(b, bool):
            return a is b or a == b
        if isinstance(a, (int, float)) and isinstance(b, (int, float)):
            return math.isclose(float(a), float(b), rel_tol=1e-9, abs_tol=1e-9)
        if isinstance(a, (list, tuple)) and isinstance(b, (list, tuple)):
            return len(a) == len(b) and all(_equal(x, y) for x, y in zip(a, b))
        if isinstance(a, dict) and isinstance(b, dict):
            return a.keys() == b.keys() and all(_equal(a[k], b[k]) for k in a)
        return a == b
    except Exception:
        return False


def _read_user(supplier):
    """Evaluate the learner's answer safely. Returns (ok, value_or_errormsg).

    `supplier` is a zero-arg lambda wrapping the learner's expression, so an
    undefined/misspelled variable surfaces here as a caught exception instead of
    crashing the cell.
    """
    try:
        return True, (supplier() if callable(supplier) else supplier)
    except Exception as e:
        return False, "%s: %s" % (type(e).__name__, e)


def check_answer(user, expected, hint, key=None, points=1):
    read_ok, user_val = _read_user(user)
    if not read_ok:
        if key is not None:
            record(key, False, points)
        print("❌ I couldn't read your answer yet (" + user_val + ").")
        print("   👉 Did you run your 🎯 cell just above this one, with no typos?")
        print("   💡 Hint:", hint)
        if key is not None:
            print("   (Truly stuck? Run  show_solution('%s')  to reveal it.)" % key)
        return False
    ok = _equal(user_val, expected)
    if key is not None:
        record(key, ok, points)
    if ok:
        print("✅ Correct! Nicely done.")
    else:
        shown = repr(user_val)
        if len(shown) > 70:
            shown = shown[:70] + "..."
        print("❌ Not yet — your result was:", shown)
        print("   💡 Hint:", hint)
        if key is not None:
            print("   (Truly stuck? Run  show_solution('%s')  to reveal it.)" % key)
    return ok


def check_quiz(user_answer, correct, explanation="", hint="", key=None, points=1):
    read_ok, val = _read_user(user_answer)
    if not read_ok:
        if key is not None:
            record(key, False, points)
        print("❌ I couldn't read your answer yet (" + val + ").")
        print("   👉 Set  answer = 'A'  (or 'B'/'C') in the cell above, then re-run.")
        if hint:
            print("   💡 Hint:", hint)
        return False
    ok = str(val).strip().upper() == str(correct).strip().upper()
    if key is not None:
        record(key, ok, points)
    if ok:
        print("✅ Correct!", explanation)
    else:
        print("❌ Not quite.")
        if hint:
            print("   💡 Hint:", hint)
        if key is not None:
            print("   (Reveal with  show_solution('%s') .)" % key)
    return ok


def show_hint(key):
    print("💡", HINTS.get(key, "No extra hint — re-read the lesson and check your variable names/types."))


def show_solution(key):
    sol = SOLUTIONS.get(key)
    if not sol:
        print("(No stored solution for this exercise.)")
    else:
        print("✅ Reference solution — try it yourself first!\n")
        print(sol)


print("✅ Self-grading engine loaded. Have fun, and don't be afraid to get ❌ — that's how you learn.")
````

### Lesson 1.1 — Max, Min & Average (the analyst's first questions)

**What it is.** Three tiny functions that answer three huge business questions:
- `max(values)` → the **largest** number ('What was our best day?')
- `min(values)` → the **smallest** number ('What was our worst day?')
- **mean / average** → the *typical* value = `sum(values) / len(values)` ('What do we do on a normal day?')

**Why it exists.** A list of 500 numbers is impossible to eyeball. These collapse a whole column of data into a single, decision-ready number.

**Real-world analogy.** Think of your monthly bank statement. You don't read all 200 transactions — you look at your *biggest* expense, your *smallest*, and your *average* daily spend. That's `max`, `min`, and mean.

**A non-code example.** A teacher with 30 test scores instantly reports the top score, the lowest score, and the class average — without adding anything up by hand for the audience.

**On the job.** When your manager asks *“What were our best and worst sales days last week, and what's the daily average?”*, this is literally the code you'd write. The average is also the foundation of nearly every metric you'll ever report.

**Worked example** — one week of store sales:

````python
weekly_sales = [1200, 980, 1450, 1610, 1305, 720, 1890]

print("Best day  :", max(weekly_sales))   # largest value
print("Worst day :", min(weekly_sales))   # smallest value

total = sum(weekly_sales)                 # add everything up
average = total / len(weekly_sales)       # divide by how many days
print("Total week:", total)
print("Daily avg :", round(average, 2))

# Python's statistics module gives the same mean without doing it by hand:
print("mean() check:", round(statistics.mean(weekly_sales), 2))
````

> 🌍 **In a real job:** The **average** can mislead when data is lopsided. If one customer spends $1,000,000 and nine spend $10, the 'average' spend is ~$100,009 — a nonsense number describing nobody. That's why analysts also use the **median** (the middle value), the **standard deviation** (how spread out the numbers are), and **percentiles**. You'll study these in statistics next; for now, just remember the average is only the *start* of the story.

### Lesson 1.2 — Growing, Shrinking & Slicing Lists

**What it is.** A list isn't frozen — you can add, insert, and remove items, and grab sub-sections ('slices'). The core moves:

| Operation | What it does |
|---|---|
| `lst.append(x)` | add `x` to the **end** |
| `lst.insert(0, x)` | add `x` to the **front** (prepend) |
| `lst.insert(i, x)` | insert `x` **at position `i`** |
| `lst.remove(x)` | remove the **first `x`** by value |
| `lst.pop()` / `lst.pop(i)` | remove & **return** the last item / item at `i` |
| `del lst[i]` | delete the item at index `i` |
| `lst[a:b]` | a **slice** — items from `a` up to (not including) `b` |
| `lst[-1]` | **negative index** — count from the end |

**Why it exists.** Real data changes: a new sale arrives, a cancelled order is removed, you need only the *top 10* rows. Lists let you model that movement.

**Real-world analogy.** A shopping cart: you `append` items as you shop, `remove` the milk you changed your mind about, and `pop` items at the checkout one by one.

**On the job.** You'll `append` results as you process records in a loop, `slice` to preview 'the first 5 rows', and use **negative indexing** to grab 'the most recent entry' — the same idea Pandas uses for `.head()` and `.tail()` later.

**Worked example** — watch the cart change step by step:

````python
cart = ["milk", "bread", "eggs"]
cart.append("butter")     # end      -> [..., "butter"]
cart.insert(0, "coffee")  # front    -> ["coffee", ...]
cart.insert(2, "sugar")   # position -> index 2 becomes "sugar"
print("after adds :", cart)

cart.remove("bread")      # by value
last = cart.pop()         # removes & returns the last item
first = cart.pop(0)       # removes & returns index 0
del cart[0]               # deletes index 0 (no return value)
print("after removes:", cart)
print("popped last :", last, "| popped first:", first)

print("first two   :", cart[:2])   # slice
print("last item   :", cart[-1])   # negative index
````

### Lesson 1.3 — Looping: Doing Something to Every Item

**What it is.** A **loop** repeats an action for each item in a collection so you don't write the same line 500 times.

- `for item in things:` → run the body **once per item** (the workhorse).
- `for i in range(n):` → loop `n` times with a counter `0, 1, ... n-1`.
- `while condition:` → keep looping **until** something becomes false.
- `enumerate(things)` → loop and get **both the position and the item**.
- `continue` → **skip** the rest of this round; `break` → **stop** the loop entirely.
- **nested loops** → a loop inside a loop (e.g., every region × every product).

**Why it exists.** Data comes in bulk. Loops are how you apply one rule — 'add up the evens', 'flag orders over \$100' — to an entire dataset automatically.

**Real-world analogy.** A cashier scanning a cart: the same action (beep, add to total) repeated for every item. `break` = a price-check halts the line; `continue` = skip a damaged item and move on.

**On the job.** Before Pandas does it for you in one line, you'll loop to compute running totals, filter records, and transform values. Understanding the loop is what makes the one-line Pandas version make sense.

**Worked example** — every looping tool in one place:

````python
prices = [9.99, 19.99, 4.50]
for p in prices:                      # for-each
    print("price:", p)

for i in range(3):                    # counted loop: 0,1,2
    print("step", i)

balance = 100
while balance > 0:                    # repeat until condition fails
    balance -= 40
print("final balance:", balance)      # 100 -> 60 -> 20 -> -20

for idx, name in enumerate(["Ann", "Bob", "Cara"]):   # position + value
    print(idx, name)

for n in range(1, 11):                # continue + break
    if n % 2 == 0:
        continue                      # skip even numbers
    if n > 7:
        break                         # stop once past 7
    print("odd kept:", n)             # 1, 3, 5, 7
````

### Lesson 1.4 — List Comprehensions (loops in one line)

**What it is.** A **list comprehension** builds a new list in a single readable line:

```python
[ expression   for item in things   if condition ]
```

Read it as: *“give me `expression` for each `item` in `things` (optionally only when `condition` is true).”*

**Why it exists.** The 'make an empty list, loop, `append`' pattern is so common that Python gives you a compact shorthand for it — less code, fewer bugs, easier to read once it clicks.

**Real-world analogy.** An assembly line with a filter: raw parts go in, each gets the *same* transformation, and a quality gate lets only the good ones through.

**On the job.** You'll use these constantly for quick data shaping: apply a 10% discount to every price, keep only orders above a threshold, convert every name to uppercase. It's the pure-Python cousin of a Pandas column operation.

**Worked example** — basic, filtered, and conditional:

````python
nums = [1, 2, 3, 4, 5]

squares = [n * n for n in nums]                 # transform every item
evens = [n for n in nums if n % 2 == 0]         # keep only some items
labels = ["hot" if t > 25 else "cold"           # choose per item
          for t in [30, 18, 27]]

print("squares:", squares)   # [1, 4, 9, 16, 25]
print("evens  :", evens)     # [2, 4]
print("labels :", labels)    # ['hot', 'cold', 'hot']
````

### Lesson 1.5 — Reversing a List (three ways)

**What it is.** Three ways to reverse order, with one important difference:

- `lst.reverse()` — reverses **in place** (changes the original list, returns `None`).
- `lst[::-1]` — a slice that returns a **new reversed list** (original untouched).
- `list(reversed(lst))` — `reversed()` gives a reverse *iterator*; wrap in `list(...)`.

**Why the difference matters.** 'In place' vs 'new copy' is one of the most common sources of bugs for beginners. If you still need the original order later, **don't** use `.reverse()` — make a reversed copy with `[::-1]`.

**Real-world analogy.** `reverse()` is flipping the actual stack of documents on your desk. `[::-1]` is photocopying them in reverse order while leaving the original stack alone.

**On the job.** You'll reverse to show **most-recent-first** timelines, or to sort then flip for a 'top to bottom' ranking. Knowing which method preserves the original saves you from silently corrupting your data.

**Worked example:**

````python
a = [1, 2, 3]
a.reverse()                       # in place
print("a after reverse():", a)    # [3, 2, 1]

b = [1, 2, 3]
b_rev = b[::-1]                   # new list
print("b:", b, "| b_rev:", b_rev) # b unchanged, b_rev reversed

c_rev = list(reversed([1, 2, 3]))
print("reversed():", c_rev)       # [3, 2, 1]
````

### Lesson 2.1 — Creating & Reading a Dictionary

**What it is.** A dictionary stores pairs: each **key** points to a **value**.

```python
product = {"name": "Laptop", "price": 1200, "in_stock": True}
```

You look things up **by key**, not by position: `product["price"]` → `1200`.

**Why it exists.** With a list of `["Laptop", 1200, True]` you'd have to *remember* that position 1 is the price. A dictionary labels each value so your code reads like plain English and never breaks when the order changes.

**Real-world analogy.** A dictionary (the book!): you don't read every page to find a word — you jump straight to it. Or your phone contacts: look up a *name* to get a *number*.

**Two ways to read a value:**
- `d["key"]` — direct, but **errors** if the key is missing.
- `d.get("key", default)` — **safe**, returns `default` (or `None`) if the key is missing.

**On the job.** You'll build a dictionary for each record you process, and use `.get()` constantly to handle missing fields without crashing your whole script.

**Worked example:**

````python
product = {"name": "Laptop", "price": 1200, "in_stock": True}

print("name       :", product["name"])            # direct access
print("price       :", product["price"])
print("color (safe):", product.get("color", "unknown"))  # missing -> default
print("has price?  :", "price" in product)         # membership test
````

### Lesson 2.2 — Adding, Updating & Removing Entries

**What it is.** Dictionaries are editable:

- `d["new"] = value` — **add** a new pair *or* **update** an existing key.
- `d["k"] += 1` — read-then-update (great for counting).
- `d.pop("k")` — **remove** key `"k"` and **return** its value.
- `del d["k"]` — delete key `"k"` (no return value).

**Why it exists.** Data is alive — inventory goes up and down, a customer cancels, a running tally grows. You need to change the dictionary as events happen.

**Real-world analogy.** A whiteboard scoreboard: you write a new team on, update a score, or wipe a team that dropped out.

**On the job.** The `d[key] += 1` pattern is the classic way to **count** occurrences (how many orders per customer, how many visits per page) before you learn the Pandas shortcut.

**Worked example:**

````python
inventory = {"apples": 50}
inventory["bananas"] = 30     # add a brand-new key
inventory["apples"] = 45      # overwrite an existing key
inventory["apples"] += 5      # read 45, add 5 -> 50

removed = inventory.pop("bananas")   # remove and capture the value
print("inventory:", inventory, "| removed:", removed)
````

### Lesson 2.3 — Looping Over a Dictionary

**What it is.** Three views let you loop over a dictionary:

- `for key in d:` (or `d.keys()`) → the **keys**.
- `for value in d.values():` → the **values**.
- `for key, value in d.items():` → **both at once** (the most useful).

`sum(d.values())` totals the numbers; `max(d, key=d.get)` finds the key with the biggest value.

**Why it exists.** A dictionary is often a whole table of results. Looping lets you total it, find the winner, or reshape it.

**Real-world analogy.** Reading a scoreboard row by row: for each team (key) you see its score (value), and you can tell who's winning.

**On the job.** *“Total sales across all regions and tell me the top region”* is a two-line answer: `sum(sales.values())` and `max(sales, key=sales.get)`.

**Worked example:**

````python
prices = {"pen": 2, "notebook": 5, "eraser": 1}

for item, cost in prices.items():     # both key and value
    print(item, "costs", cost)

print("total value :", sum(prices.values()))       # 8
print("priciest    :", max(prices, key=prices.get)) # "notebook"
````

## Module 4 · NumPy

You've been using Pandas — and most Pandas columns are backed by a **NumPy-like array** under the hood. NumPy is the number-crunching engine of the entire Python data world.

**Why it matters for the job:** when you need raw speed on numbers — feature matrices for machine learning, image pixels, simulations, or fast math across millions of values — you reach for NumPy. ### Lesson 4.1 — Arrays vs Lists (and how to create them)

**What it is.** `np.array([...])` turns a list into an **ndarray**. Key differences from a list:

- **Homogeneous:** usually one shared data type (often numeric) — which keeps it fast + compact.
- **Vectorised math:** `arr * 2` doubles every element; `list * 2` just repeats the list!
- **Multi-dimensional:** arrays can be 2D (matrices), 3D, and beyond.

Handy creators: `np.zeros(n)`, `np.ones(n)`, `np.arange(start, stop, step)`, `np.linspace(start, stop, count)`.

**Why it exists.** Python lists are flexible but slow for math. NumPy trades that flexibility for blazing numeric performance.

**Real-world analogy.** A list is a mixed drawer of odds and ends; a NumPy array is a purpose-built egg carton — uniform slots, easy to process all at once.

**On the job.** You'll create arrays for numeric features, sequences of time steps, and grids of values feeding models and charts.

**Worked example:**

````python
a = np.array([1, 2, 3, 4])
print("array   :", a, "| dtype:", a.dtype, "| shape:", a.shape)
print("zeros   :", np.zeros(3))
print("arange  :", np.arange(0, 10, 2))     # 0,2,4,6,8
print("linspace:", np.linspace(0, 1, 5))    # 0,0.25,0.5,0.75,1

print("list * 2 :", [1, 2, 3] * 2)          # repeats -> [1,2,3,1,2,3]
print("array * 2:", np.array([1, 2, 3]) * 2)  # doubles  -> [2 4 6]
````

**What it is.** Arithmetic and comparisons apply to **every element at once**:

### Lesson 4.5 — Useful Methods & Data Types

**What it is.** Arrays come with fast built-in summaries and a **dtype** (data type):

<a id="m6-concept"></a>
### Lesson 6.1 — What Machine Learning Actually Is

**The core idea.** Instead of *you* writing the rules, you show the computer lots of examples and it **learns the pattern** itself. Then it applies that pattern to new, unseen cases.

**The vocabulary (learn this):**
- **Features (X)** — the inputs/clues (e.g., house size, #bedrooms).
- **Target (y)** — what you're predicting (e.g., house price).
- **Supervised learning** — learning from examples where the answer (`y`) is known.
- **Regression** — predict a **number** (price, temperature).
- **Classification** — predict a **category** (spam / not-spam).

**The theory that keeps you honest:**
- **Train/test split** — you *train* on one slice of data and *test* on a different, unseen slice. Why? Because scoring on data the model already saw is like grading a student on the exact questions they studied — it proves nothing about the real world.
- **Overfitting** — the model *memorises* the training data (including its noise) and flops on new data. **Underfitting** — the model is too simple and misses the real pattern. Good models sit in between.

**Real-world analogy.** Studying with practice exams (train) then sitting the real exam (test). Memorising the practice answers word-for-word (overfitting) fails when the real questions differ.

### Lesson 6.3 — Your First Model: Linear Regression

**What it is.** `LinearRegression` fits the best straight line through your data. The ritual:

1. `model = LinearRegression()` — create it.
2. `model.fit(X, y)` — **learn** from the examples.
3. `model.predict(X_new)` — **forecast** new values.
4. `model.score(X, y)` — grade the fit (R², where 1.0 is perfect).

After fitting, `model.coef_` is the slope(s) and `model.intercept_` is the offset — the learned line `y = slope·x + intercept`.

**Real-world analogy.** Drawing the best-fit line through a scatter of points by eye — but the computer finds the mathematically optimal one.

**On the job.** 'Given ad spend, predict sales' or 'given size, predict price' — the classic first model, and often a surprisingly strong baseline.

**Worked example** — data that follows `y = 3x + 5`:

````python
from sklearn.linear_model import LinearRegression

X = np.array([[1], [2], [3], [4], [5]])
y = np.array([8, 11, 14, 17, 20])         # exactly 3x + 5

model = LinearRegression()
model.fit(X, y)
print("slope    :", round(model.coef_[0], 2))       # 3.0
print("intercept:", round(model.intercept_, 2))     # 5.0
print("predict 6:", round(model.predict([[6]])[0], 2))  # 23.0
print("R2 score :", model.score(X, y))              # 1.0
````

### Lesson 6.4 — Interpreting & Visualising a Regression

**What it is.** Two ways to judge a model:

- **The score (R²):** `model.score(X, y)` is **1.0** for a perfect fit and near **0** for a weak one — and it can even go **negative** if the model does worse than simply guessing the average every time. Higher is better.
- **The picture:** scatter the **actual** points and overlay the model's **predicted** line. If the line hugs the points, it's a good fit; systematic gaps reveal problems.

**Why it matters.** A single score can hide issues; the plot of *actual vs predicted* is how you sanity-check and how you explain the model to non-technical stakeholders.

**On the job.** You'll pair 'R² = 0.87' with an actual-vs-predicted chart in every model write-up — the number for rigour, the picture for trust.

**Worked example:**

````python
from sklearn.linear_model import LinearRegression

X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2.1, 3.9, 6.2, 7.8, 10.1])   # noisy, roughly 2x
model = LinearRegression().fit(X, y)
pred = model.predict(X)

fig, ax = plt.subplots()
ax.scatter(X.ravel(), y, label="Actual")
ax.plot(X.ravel(), pred, color="red", label="Predicted")
ax.set_title(f"Actual vs Predicted (R2 = {model.score(X, y):.3f})")
ax.set_xlabel("X"); ax.set_ylabel("y"); ax.legend()
plt.show()
````

### Lesson 6.7 — Experiment: Watch Overfitting Happen

**Machine learning is empirical** — you *try* things and *observe*. Here's a knob to play with: a decision tree's **`max_depth`**. A deeper tree can fit the training data more and more perfectly… which is exactly how **overfitting** creeps in.

**What to watch.** As depth increases, **training** accuracy climbs toward 100% — but that doesn't mean it generalises. A model that aces the training set yet flops on the test set is the textbook symptom of overfitting.

**On the job.** This 'turn a knob, watch train vs test' loop is the essence of model tuning. Learning to *distrust* a suspiciously perfect training score is a hallmark of a good analyst.

**Worked example — deeper trees memorise more:**

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification

Xc, yc = make_classification(n_samples=100, n_features=4, random_state=42)
for depth in [1, 3, 10]:
    m = DecisionTreeClassifier(max_depth=depth, random_state=42).fit(Xc, yc)
    print(f"max_depth={depth:>2}: training accuracy = {m.score(Xc, yc):.2f}")
````

### What a Data Pipeline Actually Is

**Plain definition:** a **data pipeline** is a *repeatable, automated path that data follows from where it's created to where it becomes useful.* Instead of a human copying spreadsheets around by hand every morning, a pipeline does it reliably, on schedule, forever.

**Real-world analogy — a city water system.** Water starts at a reservoir (**source**), flows through pipes to a treatment plant that filters out the junk (**transform/clean**), gets stored in a tank (**load/store**), and finally comes out of your tap ready to drink (**dashboard/model/report**). A data pipeline is the same idea, but for information.

**The typical stages (learn these words — you'll hear them in interviews):**

1. **Source** — where data is born: apps, websites, sensors, spreadsheets, databases, APIs.
2. **Extract** — pull the raw data out of those sources.
3. **Transform / Clean** — fix types, remove duplicates, standardise messy values (exactly the Pandas cleaning you learned in Module 3!).
4. **Load / Store** — save the clean result somewhere queryable (usually a database or 'data warehouse').
5. **Schedule** — run the whole thing automatically (e.g., every night at 2am).
6. **Validate & Monitor** — automatically check the data looks right and alert someone if it breaks.
7. **Serve** — feed the clean data into dashboards, reports, and models people actually use.

**ETL vs ELT (a common interview term, demystified):**
- **ETL** = Extract → **Transform** → Load. You clean the data *before* storing it.
- **ELT** = Extract → Load → **Transform**. You store the raw data first, then clean it *inside* the database. Modern cloud tools made ELT popular.

You don't need to *build* pipelines to be an analyst — but knowing this is where your clean data comes from (and being able to say 'ETL' with confidence) makes you far more effective and credible.

> 🌍 **In a real job:** The Pandas `.str.strip().str.lower()`, `dropna()`, `fillna()`, and `merge()` you practised in Module 3 are the *exact* operations that run inside the **Transform** stage of real pipelines — just scheduled to run automatically on millions of rows instead of by hand.
